```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv("./Files/tmdb_5000_movies.csv")
credits = pd.read_csv("./Files/tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
merged = movies.merge(credits, on="title").drop("id", axis=1)
```

```python
merged.head(1)
```

```python
# Columns to keep = ["movie_id", "genres", "keywords", "title", "overview", "cast", "crew"]

merged = merged[["movie_id", "genres", "keywords", "title", "overview", "cast", "crew"]]
```

```python
merged.duplicated().sum()
```

```python
merged.isna().sum()
```

```python
merged.dropna(inplace=True)
```

```python
def convertToList(obj):
    lst = list()
    for i in ast.literal_eval(obj):
        lst.append(i['name'])
    return lst
```

```python
merged['genres'] = merged['genres'].apply(convertToList)
```

```python
def convertKeywords(obj):
    lst = list()
    for i in ast.literal_eval(obj):
        lst.append(i['name'])
    return lst
```

```python
merged['keywords'] = merged['keywords'].apply(convertKeywords)
```

```python
def convertCast(obj):
    lst1 = ast.literal_eval(obj)
    lst2 = list()
    for i in range(min(3, len(lst1))):
        lst2.append(lst1[i]['character'])
    return lst2
```

```python
merged['cast'] = merged['cast'].apply(convertCast)
```

```python
def convertCrew(obj):
    lst1 = ast.literal_eval(obj)
    lst2 = list()
    for i in lst1:
        if i['job'] == 'Director':
            lst2.append(i['name'])
            break
    return lst2
```

```python
merged['crew'] = merged['crew'].apply(convertCrew)
```

```python
merged['overview'] = merged['overview'].apply(lambda x: x.split())
```

```python
merged.head(1)
```

```python
merged.sample(10)
```

```python
merged['genres'] = merged['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
merged['keywords'] = merged['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
merged['overview'] = merged['overview'].apply(lambda x: [i.replace(" ", "") for i in x])
merged['cast'] = merged['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
merged['crew'] = merged['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
merged.sample(5)
```

```python
merged['tags'] = merged['overview'] + merged['genres'] + merged['keywords'] + merged['cast'] + merged['crew']
```

```python
df = merged[['movie_id', 'title', 'tags']]
df.sample(5)
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df.sample(5)
```

**Appling Steming of the Tags Column to replace [Love, Loving, Lover] -> [Love, Love, Love]**

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def steming(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(steming)
```

**Converting the Tags in Text vectorization b using Bag of words of technique.**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommendMovie(movie: str):
    index = df[df['title'] == movie].index[0]
    distanceArr = list(enumerate(similarity[index]))
    sort = sorted(distanceArr, reverse=True, key=lambda x: x[1])[0:6]
    resId = [(df['title'][i[0]], movies['homepage'][i[0]]) for i in sort]
    return resId
```

```python
movie = input("Enter Movie Name:").title()
res = recommendMovie(movie)
res
```

```python
import pickle

with open('dataset_and_similarity.pkl', 'wb') as file:
    pickle.dump((df, similarity, movies), file)   
```
